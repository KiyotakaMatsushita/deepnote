```
import pandas as pd
import os
```

```
data_path = "http://api.e-stat.go.jp/rest/3.0/app/getSimpleStatsData?appId={appId}&lang=J&statsDataId=0003286682&metaGetFlg=Y&cntGetFlg=N&explanationGetFlg=Y&annotationGetFlg=Y&sectionHeaderFlg=1&replaceSpChars=0".format(appId=os.environ["E_STAT_APPLICATION_ID"])
df = pd.read_csv(data_path, skiprows=26)
```

```
df.columns
```

```
choices = df["人権相談の種類別"].unique()
print("choices: ",choices)
targets = choices[5:10]
print("targets: " ,targets)
```

```
df["is_target"] = df.apply(lambda row: row["人権相談の種類別"] in targets, axis=1 )
print(df[(df["is_target"] == True)])
```

```
df_total = df[(df["is_target"] == True) & (df["法務局・地方法務局別"] == "総数")]
df_total
```
